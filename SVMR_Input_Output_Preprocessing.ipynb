```python
import pandas as pd
```

```python
dataset=pd.read_csv('insurance_pre.csv')
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset
```

```python
dataset.columns
```

```python
independent=dataset[['age', 'bmi', 'children',  'sex_male', 'smoker_yes']]
```

```python
independent
```

```python
dependent=dataset[['charges']]
dependent
```

```python
from sklearn.model_selection import train_test_split
X_train,X_test,y_train,y_test=train_test_split(independent,dependent,test_size=0.3,random_state=0)
```

```python
X_train
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
X_train=sc.fit_transform(X_train)
X_test=sc.transform(X_test)
```

```python
X_train
```

```python
scy=StandardScaler()
y_train=scy.fit_transform(y_train)
y_test=scy.transform(y_test)
y_train
```

```python
from sklearn.svm import SVR
regressor=SVR(kernel="rbf",C=10,gamma='auto',shrinking=False,max_iter=100,degree=2)
regressor.fit(X_train,y_train)
```

```python
regressor.intercept_
```

```python
regressor.support_
```

```python
y_pred=regressor.predict(X_test)
```

```python
from sklearn.metrics import r2_score
r_score=r2_score(y_test,y_pred)
r_score
```

```python
import pickle
filename='finalized_model_MLAssignment_SVMRIPOPDataPreprocess.sav'
pickle.dump(regressor,open(filename,'wb'))
```

```python
preinput=sc.transform([[50,30.970,3,1,0]])
preinput
```

```python
loaded_model=pickle.load(open('finalized_model_MLAssignment_SVMRIPOPDataPreprocess.sav','rb'))
result=loaded_model.predict(preinput)
result
```

```python
preoutput=scy.inverse_transform([result])
```
